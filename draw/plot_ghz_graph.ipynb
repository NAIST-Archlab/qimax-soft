```python
import matplotlib.pyplot as plt
import numpy as np
# Read the file
ghz = '../data/qml_cpu/times_ghz_qimaxv1.txt'
graph = '../data/qml_cpu/times_graph_qimaxv1.txt'

fig, axs = plt.subplots(1, 1, figsize=(5, 2))
ghz = np.loadtxt(ghz, delimiter=',', skiprows=1)
graph = np.loadtxt(graph, delimiter=',', skiprows=1)
plt.plot(range(3, 1000), ghz, label = 'GHZ circuit')
plt.plot(range(4, 1000), graph, label = 'Graph circuit')
plt.xlabel('#Qubits')
plt.ylabel('Time (s)')
plt.legend()
plt.xticks([2, 200,400, 600, 800, 1000])
# plt.savefig('ghz_graph.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
